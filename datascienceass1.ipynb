```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ['07/2019', '08/2019', '09/2019', '10/2019', '11/2019']
searches = [50, 53, 59, 56, 62]
direct = [39, 47, 42, 51, 51]
social_media = [70, 80, 90, 87, 92]

x = np.arange(len(months))  # the label locations
width = 0.25  # width of the bars

# Plotting the bars
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width, searches, width, label='Searches', color='dodgerblue')
rects2 = ax.bar(x, direct, width, label='Direct', color='hotpink')
rects3 = ax.bar(x + width, social_media, width, label='Social Media', color='orange')

# Labels and titles
ax.set_ylabel('Visitors in thousands')
ax.set_title('Visitors by web traffic source')
ax.set_xticks(x)
ax.set_xticklabels(months)
ax.legend()

# Show bar values
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.tight_layout()
plt.show()
```
